```python
import pandas as pd
from tensorflow.keras.optimizers import SGD
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('shuffledLabeledSQJtrain.csv')
```

```python
X = pd.get_dummies(df.drop(['Proper SQJ'], axis=1), dtype=float)
y = df['Proper SQJ'].apply(lambda x: 1 if x == 'Yes' else 0)                    
```

```python
X_train, y_train = X, y
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(units=40, activation='sigmoid', input_dim=len(X_train.columns)))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer=SGD(learning_rate=0.01), metrics=['accuracy'])
```

```python
model.fit (X_train, y_train, epochs=40, batch_size=8)
```

```python
y_hat = model.predict(X)
y_hat = [0 if val <0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
accuracy_score(y, y_hat)
```

```python
model.save('SQJ0.01.keras')
```
